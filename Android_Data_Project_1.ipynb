The Legacy of The Moon Mission & Outer Planet Astronomy

The above two graphs may be illuminating for anyone who has wondered about the history of solar system research over the past half century. How has data collection progressed? Has it ever digressed? Has the scope of research increased, or more the amount of data looking at the same objects. Here we can see, in two breif story telling charts, how indeed the amount of data collected has grown overall though it dropped off sharply after all space programs petered out in the early 1970's along with the "moon shot." 

In the scatterplot we can see how scope of data collection has more plateaued after a recovery rather than grown appreciably.
